```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
page = requests.get('https://www.w3resource.com/python-exercises/python-basic-exercises.php')
soup = bs(page.text, 'html.parser')
```

```python
result = []
for i in soup.article.find_all('p'):
    for j in i.find_all('strong'):
        result.append(j.next_sibling)
```

```python
result
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(result)
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data = data.iloc[:151]
```
