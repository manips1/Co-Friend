```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
def w3_crawler(link):
    page = requests.get(link)
    soup = bs(page.text, 'html.parser')
    result = []
    for i in soup.article.find_all('p'):
        for j in i.find_all('strong'):
            result.append(j.next_sibling)
    
    return pd.DataFrame(result)
```

```python
data = w3_crawler('https://www.w3resource.com/python-exercises/python-basic-exercises.php')
```
